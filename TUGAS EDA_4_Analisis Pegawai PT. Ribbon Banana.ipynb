```python
# Menentukan asal pendidikan karyawan
pendidikan = df2['Pendidikan'].value_counts()
pendidikan
```

```python
plt.figure(figsize=(8, 6))
plt.pie(pendidikan, labels=pendidikan.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Agar lingkaran tampak lebih bulat
plt.title('Persentase Tingkat Pendidikan Karyawan')
plt.show()
```

Berdasarkan grafik diatas sebesar 60% dari total karyawan yang berpendidikan akhir SMA.

```python
tahun = df2['Tahun Masuk'].value_counts()
tahun =pd.DataFrame({'Tahun Masuk': tahun.index, 'Jumlah Karyawan': tahun.values})
tahun
```

```python
plt.figure(figsize=(10, 6))
plt.bar(tahun['Tahun Masuk'], tahun['Jumlah Karyawan'])
plt.xlabel('Tahun Masuk')
plt.ylabel('Jumlah Karyawan')
plt.title('Banyaknya Karyawan yang Masuk Berdasarkan Tahun')
# plt.xticks(rotation=45)  # Rotasi label sumbu x jika diperlukan untuk keterbacaan yang lebih baik
plt.show
```

Pada hasil visualisasi diatas, menunjukkan bahawa jumlah karyawan di tahun 1993 mengalami penurunan, tetapi di tahun 1994 mengalami peningkatan jumlah karyawan secara drastis.

```python
df_gby = df2.groupby('Posisi')['Pendidikan'].count()
df_gby = pd.DataFrame({'Posisi':df_gby.index, 'Pendidikan':df_gby.values})
df_gby
```

```python
# Menghitung jumlah pegawai untuk setiap kombinasi Pendidikan-Posisi
grouped_data = df.groupby(['Posisi', 'Pendidikan']).size().unstack()

# Membuat visualisasi grafik batang
grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))

# Menambahkan label dan judul
plt.xlabel('Pendidikan')
plt.ylabel('Jumlah Pegawai')
plt.title('Jumlah Pegawai Berdasarkan Pendidikan dan Posisi')

# Menampilkan legenda
plt.legend(title='Posisi')

plt.show()
```

Dari grafik diatas, disimpulkan bahwa semua pegawai dengan jabatan karyawan berpendidikan terakhir SMA, sedangkan pegawai dengan jabatan supervisor semuanya dari lulusan akademi/universitas.

```python
# Menghitung jumlah karyawan untuk setiap kombinasi Pendidikan-Posisi
grouped_data = df.groupby(['Tahun Masuk', 'Pendidikan']).size().unstack()

# Membuat visualisasi grafik batang
grouped_data.plot(kind='bar', stacked=True, figsize=(10, 6))

# Menambahkan label dan judul
plt.xlabel('Pendidikan')
plt.ylabel('Jumlah Pegawai')
plt.title('Jumlah Pegawai Berdasarkan Pendidikan dan Posisi')

# Menampilkan legenda
plt.legend(title='Posisi')

plt.show()
```

Dari grafik diatas, disimpulkan bahwa paling banyak pegawai dari lulusan SMA ditahun 1994.

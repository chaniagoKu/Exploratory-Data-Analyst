```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import statistics as stc
```

```python
dfw = pd.read_excel('FILE_LATIHAN/data_lama_belajar.xlsx')
df = dfw.dropna()
df
```

```python
df['Jumlah jam belajar(Jam per hari)'] = df['Jumlah jam belajar(Jam per hari)'].str.split().str[0]
df
```

```python
df['Jumlah jam belajar(Jam per hari)'] = df['Jumlah jam belajar(Jam per hari)'].str.replace(r'\D', '.').astype(float)
df
```

### Usia

```python
# min & max
print('Nilai minimal dari kolom Usia :', df['Usia'].min())
print('Nilai maksimal dari kolom Usia :', df['Usia'].max())

# mean/rata-rata
mean = df['Usia'].sum()/df['Usia'].count()
print('Nilai rata-rata dari kolom Usia :', mean)

# median
median = df['Usia'].count()/2
print('Nilai median dari kolom Usia', median, 'dengan nilai 20')
df[df['id_number']==17]

# modus
modus = df['Usia'].mode()
print('Nilai modus dari kolom Usia :', modus)

# varians
data = df['Usia']
list_varian = []
for bilangan in data:
    list_varian.append((bilangan - mean) ** 2)
varian = sum(list_varian) / (len(data) - 1)
print('Nilai varians dari kolom Usia :', varian)

# standar deviasi
std = stc.sqrt(varian)
print('Nilai std dari kolom Usia :', std)
```

### Waktu Belajar (Jam)

```python
# min & max
print('Nilai minimal dari kolom Jumlah jam belajar(Jam per hari) :', df['Jumlah jam belajar(Jam per hari)'].min())
print('Nilai maksimal dari kolom Jumlah jam belajar(Jam per hari) :', df['Jumlah jam belajar(Jam per hari)'].max())

# mean/rata-rata
mean = df['Jumlah jam belajar(Jam per hari)'].sum()/df['Jumlah jam belajar(Jam per hari)'].count()
print('Nilai rata-rata dari kolom Jumlah jam belajar(Jam per hari) :', mean)

# median
median = df['Jumlah jam belajar(Jam per hari)'].count()/2
print('Nilai median dari kolom Jumlah jam belajar(Jam per hari) :', median, 'dengan nilai 4')
df[df['id_number']==17]

# varians
data = df['Jumlah jam belajar(Jam per hari)']
list_varian = []
for bilangan in data:
    list_varian.append((bilangan - mean) ** 2)
varian = sum(list_varian) / (len(data) - 1)
print('Nilai varians dari kolom Jumlah jam belajar(Jam per hari) :', varian)

# standar deviasi
std = stc.sqrt(varian)
print('Nilai std dari kolom Jumlah jam belajar(Jam per hari) :', std)
```

```python
df['kategori']=df['IPK Terakhir']

for a in range(len(df['kategori'])):
    if df['kategori'][a]>3.5:
        df['kategori'][a]='tinggi'
    else:
        df['kategori'][a]='sedang'

df.groupby('kategori')['Jumlah jam belajar(Jam per hari)'].mean()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('FILE_LATIHAN/data diabetes BMI.xlsx')
df
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
integer_encoded = label_encoder.fit_transform(df['gender(x1)'])
print(integer_encoded)
df['gender(x1)']=integer_encoded
```

## Melihat komposisi berdasarkan gender

```python
sns.countplot(df['kategori BMI(x2)'])
```

```python
pivot = df.pivot_table(df, index='kategori BMI(x2)', columns='gender(x1)', aggfunc={'gender(x1)':len})
pivot.plot(kind='bar')
```

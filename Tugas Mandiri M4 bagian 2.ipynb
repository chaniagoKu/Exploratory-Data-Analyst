#### kategori BMI
Body Mass Index atau BMI adalah suatu ukuran yang digunakan untuk menunjukkan kategori berat badan seseorang.
Saat BMI seseorang menyentuh angka 23–24,9, maka ia dikategorikan mengalami kelebihan berat badan. BMI normal berada di kisaran angka 18,5–22,9. Jika seseorang memiliki BMI di bawah angka 18,5, maka ia memiliki berat badan di bawah normal.

#### Ratio HDL = perbandingan kolesterol baik
Rasio kolesterol total/ HDL merupakan variabel lipoprotein terbaik dalam memprediksi risiko penyakit kardiovaskular. Peningkatan risiko penyakit kardiovaskular dapat disebabkan oleh tingginya kadar kolesterol serum.

Tingkat HDL minimal 60 mg/dL atau lebih dapat membantu mengurangi risiko penyakit jantung. Sebaliknya, tingkat HDL kurang dari 40 mg/dL justru menaikkan risiko penyakit jantung.

#### glyhb = gula darah
Normalnya, kadar HbA1c dalam darah berkisar antara 4 hingga 5,6 %. Kadar HbA1c yang berada antara 5,7 hingga 6,4 % mengindikasikan Anda berisiko lebih tinggi mengalami diabetes. Sedangkan, kadar HbA1c yang mencapai 6,5 % atau lebih bisa berarti bahwa Anda sudah terkena diabetes.

## Komposisi jenis kelamin

```python
print(df.groupby('gender(x1)')['gender(x1)'].count())
df.groupby('gender(x1)')['gender(x1)'].count().plot.pie(autopct="%.1f%%")
plt.title("Perbandingan Jenis Kelamin")
```

## USIA

```python
#RAta Rata dan Median Usia
print("================================ \n STATISTIKA DESKRIPTIF USIA \n================================")

mean = df['Age(x3)'].mean()
print("Rata-rata :", mean)

median = df['Age(x3)'].median()
print("median :", median)
```

```python
new = df['Age(x3)']

# mencari mean, median dan modus
mean = new.mean()
median = new.median()
modus = new.mode()

# menampilkan ukuran nilai tengah
print('Mean: ',mean,'\nMedian: ',median,'\nModus: ',modus[0])

# membuat grafik
plt.figure(figsize=(15,5))
plt.hist(new,bins=35,color='grey')
plt.axvline(mean,color='red',label='Mean')
plt.axvline(median,color='yellow',label='Median')
plt.axvline(modus[0],color='green',label='Modus')
plt.xlabel('nilai_rasio')
plt.ylabel('frekuensi')
plt.title('Histogram Frekuensi Usia')

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.distplot((df[df['gender(x1)']=='female'])['Age(x3)'], color='red')
sns.distplot((df[df['gender(x1)']=='male'])['Age(x3)'], color='blue')
```

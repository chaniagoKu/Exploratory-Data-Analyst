## Daerah Rayon mana dengan rata-rata Tinggi genangan air tertinggi?

```python
# mengurutkan data berdasarkan kolom Luas genangan
sort_tinggi = df.sort_values(by='Tinggi genangan', ascending=False)
sort_tinggi.head(10)
```

Jika dilihat dari tabel diatas bahwa Rayon Wiyung memiliki rata-rata Tinggi genangan air tertinggi.

```python
# melakukan pengecekan kembali dengan menggunakan visualisasi 
data = df.groupby('Rayon')['Tinggi genangan'].sum().plot(kind='bar')
```

```python
# cek Rayon Tandes mengapa bisa tertinggi pada grafik?
h = df[df['Rayon']=='Rayon Tandes']
h.sort_values(by='Tinggi genangan', ascending=False)
```

```python
# cek rata-rata untuk mendapatkan hasil yang pasti antara Rayon Tandes dan Rayon Wiyung berdasarkan tinggi genangan air tertinggi
tandes = df[df['Rayon']=='Rayon Tandes']
wiyung = df[df['Rayon']=='Rayon Wiyung']

mean_tandes = tandes['Tinggi genangan'].mean()
print('Rata-rata tinggi genangan air Rayon Tandes :',mean_tandes)

mean_wiyung = wiyung['Tinggi genangan'].mean()
print('Rata-rata tinggi genangan air Rayon Wiyung :',mean_wiyung)
```

Berdasarkan hasil visualisasi dan tabel diatas, terlihat bahwa Rayon Tandes memiliki peringkat tertinggi karena memiliki jumlah yang lebih banyak dibandingkan dengan Rayon Wiyung. Akan tetapi, berdasarkan rata-rata tinggi genangan air tertinggi yaitu tetap dimiliki oleh Rayon Wiyung.

## Daerah Rayon mana dengan rata-rata Lama genangan air paling lama?

```python
# mengurutkan data berdasarkan kolom Lama genangan
sort_lama = df.sort_values(by='Lama genangan', ascending=False)
sort_lama.head(10)
```

```python
# melakukan pengecekan dengan pivot table berdasarkan jumlah data
pivot = df.pivot_table(df, index = 'Rayon', aggfunc='count')
pivot
```

```python
# melakukan pengecekan kembali dengan menggunakan visualisasi 
data = df.groupby('Rayon')['Lama genangan'].sum().plot(kind='bar')
```

```python
# melakukan pengecekan kembali dengan menggunakan rata-rata setiap atribut
tandes = df[df['Rayon']=='Rayon Tandes']
wiyung = df[df['Rayon']=='Rayon Wiyung']
jambangan = df[df['Rayon']=='Rayon Jambangan']
gubeng = df[df['Rayon']=='Rayon Gubeng']
genteng = df[df['Rayon']=='Rayon Genteng']

mean_tandes = tandes['Lama genangan'].mean()
print('Rata-rata Lama genangan air Rayon Tandes :',mean_tandes)

mean_wiyung = wiyung['Lama genangan'].mean()
print('Rata-rata Lama genangan air Rayon Wiyung :',mean_wiyung)

mean_jambangan = jambangan['Lama genangan'].mean()
print('Rata-rata Lama genangan air Rayon Jambangan :',mean_jambangan)

mean_gubeng = gubeng['Lama genangan'].mean()
print('Rata-rata Lama genangan air Rayon Gubeng :',mean_gubeng)

mean_genteng = genteng['Lama genangan'].mean()
print('Rata-rata Lama genangan air Rayon Genteng :',mean_genteng)
```

Berdasarkan hasil visualisasi dan tabel diatas, terlihat bahwa Rayon Gubeng memiliki peringkat tertinggi karena memiliki jumlah yang lebih banyak dibandingkan dengan Rayon Genteng. Akan tetapi, berdasarkan rata-rata lama genangan air paling lama yaitu Rayon Genteng.

## Bagaimana distribusi penyebaran dari luas genangan air?

## Apakah ada hubungan antara luas dan tinggi genangan air?

```python
# korelasi hubungan
df_kor = df.corr()
df_kor
```

```python
# menggunakan heatmap
sns.heatmap(df_kor, annot=True)
```

```python
sns.regplot(data=df, y='Tinggi genangan', x='Luas genangan')
```

Hubungan antara Luas dan Tinggi genangan air yaitu -0,77 (Negatif Kuat). Berdasarkan penelitian dan hubungan yang didapatkan yaitu jika luas genangan air semakin meningkat maka tinggi genangan air akan semakin rendah.
